## 5) Resúmenes y groupby

## 6) Transformaciones

*   assign() para crear columnas nuevas (ej: tip_percent)
*   qcut() para clasificar en rangos (ej: propina baja, media, alta).



```python
df_tips = df_tips.assign(tip_porcentaje=lambda x: (x['tip']/x['total_bill'])*100)
df_tips
```

```python
df_tips.head()
```

```python
porcentaje = df_tips['tip'] / df_tips['total_bill'] * 100
porcentaje
```

```python
df_tips['tip_porcentaje'] = df_tips['tip']/df_tips['total_bill'] * 100
df_tips
```

```python
df_tips.head()
```

```python
df_tips['nivel_propina'] = pd.qcut(df_tips['tip_porcentaje'],3, labels=['baja','media','alta'])
df_tips
```

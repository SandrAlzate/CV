## 6) Transformaciones

*   assign() para crear columnas nuevas (ej: tip_percent)
*   qcut() para clasificar en rangos (ej: propina baja, media, alta).



```python
df_tips = df_tips.assign(tip_porcentaje=lambda x: (x['tip']/x['total_bill'])*100)
df_tips
```

```python
df_tips.head()
```

```python
porcentaje = df_tips['tip'] / df_tips['total_bill'] * 100
porcentaje
```

```python
df_tips['tip_porcentaje'] = df_tips['tip']/df_tips['total_bill'] * 100
df_tips
```

```python
df_tips.head()
```

```python
cut_df_tips = pd.cut(df_tips['tip_porcentaje'], bins=4)
# Mostrar los intervalos creados
print("Rangos con cut (bins=3):")
print(cut_df_tips.cat.categories)
```

```python
df_tips['nivel_propina'] = pd.qcut(df_tips['tip_porcentaje'], q=4, labels=['baja','media','alta','Muy alta'])
df_tips
```

```python
qcut_q = pd.qcut(df_tips["tip_porcentaje"], q=4)
print("\nRangos con qcut (q=4):")
print(qcut_q.cat.categories)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.boxplot(x=df_tips["tip_porcentaje"], color="lightblue")


q1 = df_tips["tip_porcentaje"].quantile(0.25)
q2 = df_tips["tip_porcentaje"].quantile(0.50)  # mediana
q3 = df_tips["tip_porcentaje"].quantile(0.75)
max_val = df_tips["tip_porcentaje"].max()


for q, label in zip([q1, q2, q3, max_val],
                    ["Q1 (25%)", "Q2 (Mediana)", "Q3 (75%)", "Máximo (Q4)"]):
    plt.axvline(q, color="red", linestyle="--")
    plt.text(q, 0.05, label, rotation=90, verticalalignment="bottom", color="red")


plt.title("Boxplot de Tip (%) con cuartiles y máximo", fontsize=14)
plt.xlabel("Tip (%)", fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

col = "tip_porcentaje"  # cámbialo si quieres otra variable

q1 = df_tips[col].quantile(0.25)
q2 = df_tips[col].quantile(0.50)
q3 = df_tips[col].quantile(0.75)
min_val = df_tips[col].min()
max_val = df_tips[col].max()
iqr = q3 - q1
lower_whisker = max(min_val, q1 - 1.5 * iqr)
upper_whisker = min(max_val, q3 + 1.5 * iqr)

plt.figure(figsize=(10, 4))
plt.boxplot(df_tips[col], vert=False, showfliers=True)

def vline_with_label(x, label):
    plt.axvline(x, linestyle="--")
    plt.annotate(label, xy=(x, 1.05), xytext=(x, 1.18),
                 textcoords="data", ha="center", va="bottom",
                 rotation=90, arrowprops=dict(arrowstyle="->", lw=1))

vline_with_label(q1, "Q1 (25%)")
vline_with_label(q2, "Mediana (Q2, 50%)")
vline_with_label(q3, "Q3 (75%)")
vline_with_label(lower_whisker, "Bigote inferior")
vline_with_label(upper_whisker, "Bigote superior")
vline_with_label(max_val, "Máximo (posible outlier)")

plt.title(f"Boxplot anotado de {col}")
plt.xlabel("Valor")
plt.tight_layout()
plt.show()
```

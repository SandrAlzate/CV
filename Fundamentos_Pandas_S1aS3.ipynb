## üîπ SESI√ìN 1: Introducci√≥n a Pandas ‚Äì Series y DataFrames


```python
# Importar pandas
import pandas as pd
pd.__version__
```


### Series
Una Serie es una estructura **unidimensional** con √≠ndices.


```python
# Crear una Serie con √≠ndice personalizado
serie = pd.Series([10, 20, 30, 40], index=["a", "b", "c", "d"], name="valores")
```

```python
print(serie)
```

```python
print("Acceso por etiqueta:", serie["b"])
```

```python
print("Acceso por posici√≥n:", serie.iloc[2])
```


### DataFrame
Un DataFrame es una tabla bidimensional con filas y columnas.


```python
# Crear un DataFrame desde un diccionario
data = {
    "Nombre": ["Ana", "Luis", "Mar√≠a"],
    "Edad": [20, 22, 19],
    "Carrera": ["Ingenier√≠a", "Matem√°ticas", "Biolog√≠a"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Acceso b√°sico
print("Columna 'Nombre':\n", df["Nombre"])
```

```python
print("\nFila con .loc (etiqueta 0):\n", df.loc[0])
```

```python
print("\nFila con .iloc (posici√≥n 1):\n", df.iloc[1])
```

## üîπ SESI√ìN 2: Importaci√≥n y exportaci√≥n de datos (CSV / Excel)


```python
# Crear un CSV de ejemplo
ejemplo = pd.DataFrame({
    "nombre": ["Ana","Luis","Mar√≠a"],
    "nota": [4.5, 3.8, 4.2],
    "fecha": pd.to_datetime(["2025-08-01","2025-08-02","2025-08-03"])
})
csv_path = "estudiantes.csv"
ejemplo.to_csv(csv_path, index=False)
print("Archivo guardado:", csv_path)
```

```python
# Cargar el CSV
df_csv = pd.read_csv("estudiantes.csv", parse_dates=["fecha"])
df_csv.head()
```

```python
# Crear y cargar un Excel de ejemplo
xlsx_path = "ventas.xlsx"
with pd.ExcelWriter(xlsx_path) as writer:
    pd.DataFrame({"producto":["A","B","C"], "precio":[10,20,30]}).to_excel(writer, sheet_name="Hoja1", index=False)
    pd.DataFrame({"producto":["A","B","C"], "unidades":[5,7,2]}).to_excel(writer, sheet_name="Hoja2", index=False)

df_xlsx = pd.read_excel(xlsx_path, sheet_name="Hoja1")
df_xlsx.head()
```

## üîπ SESI√ìN 3: Indexaci√≥n, Valores faltantes y Nuevas Variables



### Indexaci√≥n b√°sica y avanzada
- `.loc[fila_label, col_label]` ‚Üí por **etiqueta**  
- `.iloc[fila_pos, col_pos]` ‚Üí por **posici√≥n**


```python
# DataFrame de ejemplo
dfm = pd.DataFrame({
    "id":[1,2,3,4,5],
    "producto":["A","B","C","D","E"],
    "precio":[10.0, 12.5, None, 15.0, 9.5],
    "unidades":[5, 7, 2, None, 4]
}).set_index("id")
dfm
```

```python
# Acceso con loc/iloc
print("loc filas 2..4, columnas 'producto' y 'precio'")
print(dfm.loc[2:4, ["producto","precio"]])

print("\niloc filas 0..2, columnas 0..1")
print(dfm.iloc[0:3, 0:2])
```


### Manejo de valores faltantes
- `df.isnull()` ‚Üí detecta nulos  
- `df.dropna()` ‚Üí elimina filas con nulos  
- `df.fillna(valor)` ‚Üí reemplaza valores nulos


```python
dfm.isnull()
```

```python
# Eliminar valores faltantes
dfm_drop = dfm.dropna()
dfm_drop
```

```python
dfm.describe()
```

```python
# Reemplazar valores faltantes
dfm_fill = dfm.fillna({"precio": dfm["precio"].mean(), "unidades": 0})
dfm_fill
```


### Creaci√≥n de nuevas variables


```python
# Nueva columna: ingreso total
dfm["ingreso_total"] = dfm["precio"] * dfm["unidades"]
dfm
```

```python
# Nueva columna categ√≥rica
dfm["categoria"] = ["alto" if x>=50 else "bajo" for x in dfm["ingreso_total"].fillna(0)]
dfm
```
